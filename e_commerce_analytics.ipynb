```python
#revenue by status
revenue_by_status=orders_payments.groupby("status")["amount_paid"].sum()
revenue_by_status
```

```python
#combine users with orders_payments table
users_orders_payments=pd.merge(users.copy(),orders_payments.copy(),on="user_id",how="left",indicator=True,validate="one_to_many")
users_orders_payments
```

```python
#aggregate revenue per user
revenue_per_user=users_orders_payments.groupby("name",as_index=False)["amount_paid"].sum()
revenue_per_user=revenue_per_user.sort_values(by="amount_paid",ascending=False)
revenue_per_user
#John Doe paid highest amount
```

```python
#group by name and country
#calculate total sales and average saless
sales_summary=(
    users_orders_payments
    .groupby(["name","country"],as_index=False)
    .agg(
        total_sales=("amount_paid","sum"),
        average_sales=("amount_paid","mean"))
    .sort_values(by="total_sales",ascending=False)
)
#fill empty fields ie NaN with 0 if no amount exist
sales_summary["average_sales"]=sales_summary["average_sales"].fillna(0)
sales_summary
```

```python
paypal_payments=orders_payments.loc[(orders_payments["status"]=="Completed")&(orders_payments["payment_method"]=="PayPal")].copy()
paypal_payments
```

```python
#fetch users who paid via Debit Card
debit_card_payments=orders_payments.loc[(orders_payments["status"]=="Completed")&(orders_payments["payment_method"]=="Debit Card")].copy()
debit_card_payments
```

```python
sales_summary
```

```python
sales_per_country= pd.pivot_table(
    sales_summary,
    index="country",
    values=["total_sales", "average_sales"],
    aggfunc={"total_sales": "sum", "average_sales": "mean"}
)

sales_per_country=sales_per_country[["total_sales","average_sales"]]
sales_per_country
```

```python
print("Total revenue: ",sales_per_country["total_sales"].sum())
```

```python
#convert to excel sheet ie one sheet
sales_per_country.reset_index().to_excel("sales_per_country.xlsx",sheet_name="Sales per country",index=False)
```

```python
#convert to excel sheet with multiple sheets
sorted_by_total_sales=sales_per_country.sort_values(by="total_sales",ascending=False)
sorted_by_average_sales=sales_per_country.sort_values(by="average_sales",ascending=False)
```

```python
#convert to excel
with pd.ExcelWriter("revenue_per_country.xlsx",engine="openpyxl") as writer:
    # print(writer.sheets)
    sales_per_country.reset_index().to_excel(writer,sheet_name="All countries",index=False)
    by_total_sales.reset_index().to_excel(writer,sheet_name="Sorted by Total sales",index=False)
    by_average_sales.reset_index().to_excel(writer,sheet_name="Sorted by Average sales",index=False)
    #add conditional highlighting
    highlight=PatternFill(
        start_color="0000FF00",
        end_color="0000FF00",
        fill_type="solid")
    for sheet_name in ["All countries","Sorted by Total sales","Sorted by Average sales"]:
        ws=writer.sheets[sheet_name]
        max_row=ws.max_row
        ws.conditional_formatting.add(f"B2:B{max_row}",FormulaRule(formula=[f"B2=MAX(B$2:B${max_row})"],fill=highlight))
        ws.conditional_formatting.add(f"C2:C{max_row}",FormulaRule(formula=[f"C2=MAX(C$2:C${max_row})"],fill=highlight))
    
    
    
    
```
